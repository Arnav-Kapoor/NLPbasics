```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
from sklearn.model_selection import train_test_split
```

```python
input_files=['edgar_allan_poe.txt', 'robert_frost.txt']
```

```python
f=open('datasets/edgar_allan_poe.txt')
print(f.read())
```

```python
f=open('datasets/robert_frost.txt')
print(f.read())
```

```python
input_texts=[]
labels=[]
    
for label, f in enumerate(input_files):
    print(f"{f} corresponds to label {label}")

    for line in open(f):
        line=line.rstrip().lower() ##remove newline character and lowercase it
        if line:
            #remove punctuations
            line=line.translate(str.maketrans('','',string.punctuation))

            input_texts.append(line)
            labels.append(label)
```

```python
train_text,test_text,y_train,y_test=train_test_split(input_texts,labels)
```

```python
idx=1
word2idx={'<unk>':0}
```

```python
#populate
for text in train_text:
    tokens=text.split()
    for token in tokens:
        if token not in word2idx:
            word2idx[token]=idx
            idx+=1
```

```python
word2idx
```

```python
len(word2idx)
```

```python
#convert data into integer form
train_text_int=[]
test_text_int=[]

for text in train_text:
    tokens=text.split()
    line_as_int=[word2idx[token] for token in tokens]
    train_text_int.append(line_as_int)
    
for text in test_text:
    tokens=text.split()
    line_as_int=[word2idx.get(token,0) for token in tokens]
    test_text_int.append(line_as_int)
```

```python
train_text_int[100:105]
```

```python
#initialize A and pi matrices for both clasees

V=len(word2idx)

A0=np.ones((V,V))
pi0=np.ones(V)

A1=np.ones((V,V))
pi1=np.ones(V)
```

```python
#compute counts for A and pi
def compute_counts(text_as_int,A,pi):
    for tokens in text_as_int:
        last_idx=None
        for idx in tokens:
            if last_idx is None:
                #its the first word in a sentence
                pi[idx]+=1
            else:
                #last word exists, so count a transotion
                A[last_idx,idx]+=1
            #update last idx
            
            last_idx=idx
```

```python
compute_counts([t for t,y in zip(train_text_int,y_train) if y==0],A0,pi0)
compute_counts([t for t,y in zip(train_text_int,y_train) if y==1],A1,pi1)
```

```python
#normalize A and pi so they are valid probablity matrices
A0/=A0.sum(axis=1,keepdims=True)
pi0/=pi0.sum()

A1/=A1.sum(axis=1,keepdims=True)
pi1/=pi1.sum()
```

```python
#log A and pi since we dont need the actual probs
logA0=np.log(A0)
logpi0=np.log(pi0)

logA1=np.log(A1)
logpi1=np.log(pi1)
```

```python
#compute priors
count0=sum(y==0 for y in y_train)
count1=sum(y==1 for y in y_train)
total=len(y_train)
p0=count0/total
p1=count1/total
logp0=np.log(p0)
logp1=np.log(p1)
p0,p1
```

```python
#build a classifier

class Classifier:
    def __init__(self,logAs,logpis,logpriors):
        self.logAs=logAs
        self.logpis=logpis
        self.logpriors=logpriors
        self.K=len(logpriors) #number of classes
        
    def _compute_log_likelihood(self,input_,class_):
        logA=self.logAs[class_]
        logpi=self.logpis[class_]
        
        last_idx=None
        logprob=0
        for idx in input_:
            if last_idx is None:
                #its the forst token
                logprob+=logpi[idx]
            else:
                logprob+=logA[last_idx,idx]
            #update last_idx
            last_idx=idx
        return logprob
    
    def predict(self,inputs):
        predictions=np.zeros(len(inputs))
        for i, input_ in enumerate(inputs):
            posteriors=[self._compute_log_likelihood(input_,c)+self.logpriors[c] \
                       for c in range(self.K)]
            pred=np.argmax(posteriors)
            predictions[i]=pred
        return predictions
```

```python
#each array ,ust be in order since classes are assumed to index theses lists
clf=Classifier([logA0,logA1],[logpi0,logpi1],[logp0,logp1])
```

```python
P_train=clf.predict(train_text_int)
print(f"Train acc :{np.mean(P_train==y_train)}")
```

```python
P_test=clf.predict(test_text_int)
print(f"Test acc: {np.mean(P_test==y_test)}")
```
